```
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```
column_name=[' Instruction Medium ', ' Male Teachers ', ' Pre Primary Sectin Avilable ', ' Board for Class 10', ' School Type ', ' Classes ', ' Female Teacher ', ' Pre Primary Teachers ', ' Board for Class 10+2 ', 'Meal provided', ' Establishment ', ' School Area ', ' School Shifted to New Place ', ' Head Teachers ', ' Head Teacher ', ' Is School Residential ', ' Residential Type ', ' Total Teachers ', ' Contract Teachers ', ' Management ', ' Village / Town ', ' Cluster ', ' Block ', ' District ', ' State ', ' UDISE Code  ', ' Building ', ' Class Rooms ', ' Boys Toilet ', ' Girls Toilet ', ' Computer Aided Learning ', ' Electricity ', ' Wall ', ' Library ', ' Playground ', ' Books in Library ', ' Drinking Water ', ' Ramps for Disable ', ' Computers ']
```

```
page=requests.get('https://schools.org.in/Karnataka')
```

```
soup=BeautifulSoup(page.content,'lxml')
print(soup.prettify())
```

```
tables=soup.find('table',class_='table table-striped')
table_dist=tables.find_all('a')
```

```
dist_url=[]
for i in table_dist:
     dist_url.append('https://schools.org.in/'+i.get('href'))
```

```
temp=[]
for url in dist_url:
    page1=requests.get(url)
    soup1=BeautifulSoup(page1.content,'lxml')
    tables1=soup1.find('table',class_='table table-striped')
    table_block=tables1.find_all('a')
    temp.append(table_block)
```

```
block_url=[]
for h in temp:
    for i in h:
        block_url.append('https://schools.org.in/'+i.get('href'))
```

```
len(block_url)
```

```
len(block_url)
```

```
temp1=[]
for url in block_url:
    page2=requests.get(url)
    soup2=BeautifulSoup(page2.content,'lxml')
    tables3=soup2.find('table',class_='table table-striped')
    table_block2=tables3.find_all('a')
    temp1.append(table_block2)
```

```
cluster_url=[]
for h in temp1:
    for i in h:
        cluster_url.append('https://schools.org.in/'+i.get('href'))
```

```
len(cluster_url)
```

```
temp2=[]
for i in cluster_url[500:1000]:
    page3=requests.get(i)
    soup3=BeautifulSoup(page3.content,'lxml')
    tables4=soup3.find('table',class_='table table-striped')
    table_block3=tables4.find_all('a')
    temp2.append(table_block3)
```

```
len(temp2)
```

```

school_url1=[] 
for h in temp2:
    for i in h:
        school_url1.append('https://schools.org.in/'+i.get('href'))
```

```
len(school_url1)
```

```
values2=[]


for a in school_url1[6000:]:
    page_a=requests.get(a)
    soup_a=BeautifulSoup(page_a.content,'lxml')
    tables_a=soup_a.find_all('li',class_='list-group-item')
    temp_a=[]
    for b in range(len(tables_a)):
        if tables_a[b].b.string=='Meal':
            temp_a.append(tables_a[b].contents[3])
        elif tables_a[b].contents[1].string=='Residential Type':
            temp_a.append(tables_a[b].b.string)
        else:   
            temp_a.append(tables_a[b].b.string)
        
    values2.append(temp_a)
```

```
len(values2)
```

```
schools_in_karnataka4=pd.DataFrame(data=values2,columns=column_name)



```
